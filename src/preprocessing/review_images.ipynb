```python
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
import tifffile
```

```python
bone_images = []
bone_path = "/data/public/NANO_dual_energy/20211205_cirs_phantom_kvs/bone/"
for filename in os.scandir(bone_path):
    if filename.is_file():
        #print(filename.path)
        bone_images.append(np.load(filename.path))
```

```python
soft_images = []
soft_path = "/data/public/NANO_dual_energy/20211205_cirs_phantom_kvs/soft/"
for filename in os.scandir(bone_path):
    if filename.is_file():
        #print(filename.path)
        soft_images.append(np.load(filename.path))
```

```python
np.sum(bone_images[10]-soft_images[10])
```

```python
%matplotlib inline
fig, ax = plt.subplots(nrows=10, ncols=2, figsize=(9, 4*10))
for i in range(0,10):
    ax[i][0].imshow(bone_images[i], cmap='gray', vmin=-0.5, vmax=1)
    ax[i][0].set_title('Bone')
    ax[i][1].imshow(soft_images[i], cmap='gray', vmin=-0.5, vmax=1)
    ax[i][1].set_title('Soft')

fig.tight_layout()
```
